### Text cleaning

#### step 1: converting to lowercase

```python
row_doc1=[doc.lower() for doc in data['Comment']]
row_doc2=[doc.lower() for doc in data['Emotion']]
print(row_doc1)
print(row_doc2)
```

#### step 2: tokenization

```python
from nltk import word_tokenize
tokenized_data1=[word_tokenize(doc) for doc in row_doc1]
print(tokenized_data1)
tokenized_data2=[word_tokenize(doc) for doc in row_doc2]
print(tokenized_data2)
```

#### step 3: punctuation removal

```python
import string
import re
regex = re.compile('[%s]' % re.escape(string.punctuation))
tokenized_data_no_punctuation1=[]
for review in tokenized_data1:
    new_review=[]
    for token in review:
        new_token = regex.sub(u'', token)
        if not new_token==u'':
            new_review.append(new_token)

    tokenized_data_no_punctuation1.append(new_review)
print(tokenized_data_no_punctuation1)
```

#### step 4: Removing stopwords

```python
from nltk.corpus import stopwords
tokenized_data_no_stopwords1=[]
for doc in tokenized_data_no_punctuation1:
    new_term_vector=[]
    for word in doc:
        if not word in stopwords.words('english'):
            new_term_vector.append(word)
    tokenized_data_no_stopwords1.append(new_term_vector)
print(tokenized_data_no_stopwords1)
            
    
```

#### step5:stemming and lemmatisation

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
porter=PorterStemmer()
wordnet=WordNetLemmatizer()
preprocessed_data1=[]
for data in tokenized_data_no_stopwords1:
    final_data=[]
    for word in data:
        porter.stem(word)
        wordnet.lemmatize(word)
        final_data.append(word)
    preprocessed_data1.append(final_data)
print(preprocessed_data1)
    
    
        
```

### Feature extraction

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer


#converting processed documents to string
doc_string1=[' '.join(doc) for doc in preprocessed_data1]

print(doc_string1)

```

```python
#using bag of words method
bowvector=CountVectorizer()
bow_matrix1=bowvector.fit_transform(doc_string1)


# vocabulary for refrence
print('\n vocabulary of Bag of words')
print(bowvector.get_feature_names_out())

#Bag of words matrix
print('\n Bag of words matrix:\n')
print('Column:comment\n',bow_matrix1.toarray())

```

```python
#using tfidf method
tfidfvector=TfidfVectorizer()
tfidf_matrix1=tfidfvector.fit_transform(doc_string1)

#vocabulary
print('\n vocabulary of tfidf')
print(tfidfvector.get_feature_names_out())
#Tf-IDF matrix
print('\n TF-IDF matrix:\n')
print('Column:Comment\n',tfidf_matrix1.toarray())
```

### 3. Model Development

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

#### SVC(support vector machine)

```python
x=bow_matrix1.toarray()
y=data['Emotion']
np.reshape(y, (1,np.product(y.shape)))
X_train,X_test,Y_train,Y_test=train_test_split(x,y)
```

```python
from sklearn.svm import SVC
svm_model=SVC()
svm_model.fit(X_train,Y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_model=MultinomialNB()
naive_model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
nb_predict=naive_model.predict(X_test)
svm_predict=svm_model.predict(X_test)
nb_accuracy=accuracy_score(Y_test,nb_predict)
svm_accuracy=accuracy_score(Y_test,svm_predict)
nb_f1_score=f1_score(Y_test,nb_predict,average='weighted')
svm_f1_score=f1_score(Y_test,svm_predict,average='weighted')

# Print results
print("Naive Bayes Accuracy:", nb_accuracy)
print("Naive Bayes F1 Score:", nb_f1_score)
print("SVM Accuracy:", svm_accuracy)
print("SVM F1 Score:", svm_f1_score)
```

### Its clear from the above score that in this case SVM model is more accurate
